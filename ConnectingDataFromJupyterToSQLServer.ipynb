```python
# CONNECTING JUPYTER TO MY EMADEPROD DATABASE 
#IN SQL SERVER using the Directory
import pandas as pd
import urllib.parse
from sqlalchemy import create_engine

server = r'JERRY\MSSQLSERVER01'
database = 'EmadeProd'

params = urllib.parse.quote_plus(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=Yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

files = {
    "EmadeProducts": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeProducts.csv",
    "EmadeStores": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeStores.csv",
    "EmadeCustomers": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeCustomers.csv",
    "EmadeOrders": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeOrders.csv",
    "EmadeStocks": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeStocks.csv",
    "EmadeCategories": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeCategories.csv",
    "EmadeStaffs": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeStaffs.csv",
    "EmadeOrderItems": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeOrderItems.csv",
    "EmadeBrands": r"C:\Users\HP\Documents\EMADE\Python Training\EmadeBrands.csv"
}

for table_name, file_path in files.items():
    try:
        df = pd.read_csv(file_path)
        df.columns = [col.replace(' ', '_').replace('-', '_').replace('.', '_') for col in df.columns]

        df.to_sql(
            name=table_name,
            con=engine,
            if_exists='replace',
            index=False,
            schema='dbo'
        )

        print(f"{table_name} loaded successfully into {database}")

    except Exception as e:
        print(f"Error loading {table_name}: {e}")

print("All files have been processed.")
```
